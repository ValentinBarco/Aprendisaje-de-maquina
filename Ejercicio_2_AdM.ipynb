# Ejercicio 2


```
mediaA = 50;
mediaB = 30;
muestras = 50;
desvioA = 5.0;
desvioB = 1.5;

np.random.seed(1);
A = pd.Series(np.random.normal(loc= mediaA, scale=desvioA, size=muestras));

Dataset1 = pd.DataFrame({'Target':'Naranjas','Diametro':A});

B = pd.Series(np.random.normal(loc=mediaB, scale=desvioB, size=muestras));

Dataset2 = pd.DataFrame({'Target':'Mandarinas','Diametro':B});
print(Dataset1)
print(Dataset2)
```

```
sns.pairplot(Dataset1, diag_kind="kde");
sns.pairplot(Dataset2, diag_kind="kde");

Grafico = pd.concat([Dataset1,Dataset2],ignore_index= True,axis=0)

Grafico.head()

```

# Prediccion y curva Roc

```

valor_de_clasificacion_Naranjas = mediaA - 3*desvioA;
valor_de_clasificacion_Mandarinas = mediaB + 3*desvioB;

valor_de_clasificacion = 40;

Grafico['Prediccion'] = [0 if x < valor_de_clasificacion else 1 for x in Grafico['Diametro']]

Grafico['EsNaranja'] = [1 if x == 'Naranjas' else 0 for x in Grafico['Target']]

Grafico['VerdaderoPositivo'] = Grafico['Prediccion'] & Grafico['EsNaranja']


fpr, tpr, threshold = roc_curve(Grafico['VerdaderoPositivo'],Grafico['Prediccion'])

plt.plot(fpr, tpr, "ro--")

```
